```python
df2.isnull().sum()*100/len(df2) #calculating percentage of missing values . if a columns percentage is more than 70 to 75 % it becomes useless then but depends on senario as well
```

```python
# how to deal with missing values in EDA analysis ? categorical/object, Numeric (float,int), boolean
# categorical/object - replace with mode
# numeric - replace with mean/median
# boolean - replace with mode

#perform these operations on dataset df
# 1. replace missing values in embarked with mode
# 2. replace missing values in age with median
# 3. replace missing values in deck with mode
# 4. replace missing values in embark_town with mode
# 5. replace missing values in alone with mode
# 1. replace missing values in embarked with mode code it up
df2['embarked'].value_counts()
df2['embarked'].fillna('S',inplace=True)
df2['embarked'].isnull().sum()
df2['embarked'].value_counts()
df2.isnull().sum()
df2["embarked"]
```

```python
df2["age"].median()
df2["age"].fillna(df2["age"].median(),inplace=True)
df2.isnull().sum()
```

```python
df3=df2
df3.isnull().sum()
```

```python
df3.isnull().sum()
```

```python
# 4. replace missing values in embark_town with mode
df4=df3
df4.isnull().sum()
df4['embark_town'].value_counts()
df4['embark_town'].fillna(df4['embark_town'].mode()[0],inplace=True) #mode()[0] is used to get the first value of mode
df4.isnull().sum()
```
